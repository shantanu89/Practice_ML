```python
## Independant and Dependant features
```

```python
X=dataset.iloc[:,:-1]
Y=dataset.iloc[:,-1]
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,Y_train,y_test=train_test_split(X,Y,test_size=0.33,random_state=42)
```

```python
x_train.shape
```

```python
Y_train.shape 
```

```python
## Standardize or feature scaling the dataset
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
scaler
```

```python
x_train=scaler.fit_transform(x_train)
```

```python
x_test=scaler.transform(x_test) # To avoid data leakage issue we just tranform the test data
```

```python
x_train
```

```python
x_test
```

# Model Training

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
regression=LinearRegression()
```

```python
regression
```

```python
regression.fit(x_train,Y_train)
```

```python
## Print the coeficient and intercept
```

```python
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
## prediction for test data
```

```python
reg_prediction=regression.predict(x_test)
```

```python
reg_prediction
```

```python
## Assumptions of Linear Regresstion
```

```python
plt.scatter(y_test,reg_prediction)
plt.xlabel("Test Truth data")
plt.ylabel("Test Predicted data")
```

```python
## Residuals
```

```python
residuals=y_test-reg_prediction
```

```python
residuals
```

```python
sns.displot(resuduals,kind="kde")
```

```python
## scatter plot with prediction and residuals 
## Unifrom Distribution
```

```python
plt.scatter(reg_prediction,residuals)
```

```python
# Performance Matrix
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
print(mean_squared_error(y_test,reg_prediction))
print(mean_absolute_error(y_test,reg_prediction))
print(np.sqrt(mean_squared_error(y_test,reg_prediction)))
```

```python
## R square and Adjested R square
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,reg_prediction)
print(score)
```

```python
# Adjusted R square

1-(1-score)*(len(y_test)-1)/(len(y_test)-x_test.shape[1]-1)
```

```python
#Adjusted square should be less then r square
```
